<h3>Scatter plot(Numerical-Numerical)</h3>

```python
#let's plot scatter plot for tips dataset columns
#appling hue is called doing multivariate analysis because we are dealing with more than two columns
sns.scatterplot(x="total_bill",y="tip",data = tips,hue = "sex",style = "smoker",size = tips["size"])
```

<h3>Bar plot (Numerical-Categorical)</h3>

```python
titanic.head()
```

```python
#yaxis is average age
sns.barplot(x="Pclass",y="Age",data = titanic)
```

```python
# sns.barplot(x="Pclass",y="Fare",data = titanic,hue="Sex")
sns.barplot(x="Pclass",y="Fare",data = titanic,hue="Survived")
```

<h3>Box plot(Numerical-Categorical)</h3>

```python
sns.boxplot(x="Sex",y="Age",data=titanic,hue = "Survived")
```

<h3>Dist plot(Numerical-Categorical)</h3>

```python
sns.distplot(x=titanic[titanic["Survived"]==0]["Age"],hist=False)
sns.distplot(x=titanic[titanic["Survived"]==1]["Age"],hist=False)
```

```python
# from above graph we can say that if age is small then probability of survival is high and at the age of 30 
# around the probability of survival is lower
```

<h3>HeatMap(Categorical-Categorical)</h3>

```python
#we want to know that how many people survived in each pclass
```

```python
pd.crosstab(titanic["Pclass"],titanic["Survived"])
```

```python
# let's plot graph of this data
sns.heatmap(pd.crosstab(titanic["Pclass"],titanic["Survived"]))
```

```python
(titanic.groupby("Pclass")["Survived"].mean()*100)
```

```python
(titanic.groupby("Sex")["Survived"].mean()*100)
```

```python
(titanic.groupby("Embarked")["Survived"].mean()*100)
```

<h3>ClusterMap(Categorical-Categorical)</h3>

```python
sns.clustermap(pd.crosstab(titanic["Parch"],titanic["Survived"]))
```
